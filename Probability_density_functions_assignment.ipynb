```python
import pandas as pd
import numpy as np
import math
```

```python
data=pd.read_csv("data.csv",encoding="latin1")
```

```python
x=data["no2"].dropna().values
```

```python
r=102303327
```

```python
a_r=0.05*(r%7)
b_r=0.3*((r%5)+1)
```

```python
z=x+a_r*np.sin(b_r*x)
```

```python
mu=np.mean(z)
variance=np.mean((z-mu)**2)
lam=1/(2*variance)
c=math.sqrt(lam/math.pi)
```

```python
print("mu =",mu)
print("lambda =",lam)
print("c =",c)
```

```python
import matplotlib.pyplot as plt

z_line=np.linspace(min(z),max(z),300)
pdf=c*np.exp(-lam*(z_line-mu)**2)

plt.plot(z_line,pdf)
plt.xlabel("z")
plt.ylabel("Probability Density")
plt.title("Estimated Probability Density Function")
plt.show()
```
